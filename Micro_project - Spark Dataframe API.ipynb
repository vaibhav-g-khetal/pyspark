```python
from pyspark.sql.functions import *
```

```python
file_path = "/databricks-datasets/learning-spark-v2/sf-fire/sf-fire-calls.csv"

raw_sf_fire_calls_df = spark.read.format("csv")\
                            .option("inferSchema","true")\
                            .option("header","true")\
                            .load(file_path)

display(raw_sf_fire_calls_df)
```

```python
raw_sf_fire_calls_df.printSchema()
```

```python
# Removing spaces from columns names.
# CallDate, WatchDate have a str data type need to change into date format plus AvailableDtTm need to convert into timestamp format


cleaned_sf_fire_calls_df = raw_sf_fire_calls_df.withColumnRenamed("Call Number", "CallNumber")\
                                               .withColumnRenamed("Unit ID", "UnitID")\
                                               .withColumnRenamed("Incident Number", "IncidentNumber")\
                                               .withColumnRenamed("Call Date", "CallDate").withColumn("CallDate", to_date("CallDate", "MM/dd/yyyy"))\
                                               .withColumnRenamed("Watch Date", "WatchDate").withColumn("WatchDate", to_date("WatchDate", "MM/dd/yyyy"))\
                                               .withColumnRenamed("Call Final Disposition", "CallFinalDisposition")\
                                               .withColumnRenamed("Available DtTm", "AvailableDtTm").withColumn("AvailableDtTm", to_timestamp("AvailableDtTm",                                                                     "MM/dd/yyyy hh:mm:ss a"))\
                                               .withColumnRenamed("Zipcode of Incident", "Zipcode")\
                                               .withColumnRenamed("Station Area", "StationArea")\
                                               .withColumnRenamed("Final Priority", "FinalPriority")\
                                               .withColumnRenamed("ALS Unit", "ALSUnit")\
                                               .withColumnRenamed("Call Type Group", "CallTypeGroup")\
                                               .withColumnRenamed("Unit sequence in call dispatch", "Unitsequenceincalldispatch")\
                                               .withColumnRenamed("Fire Prevention District", "FirePreventionDistrict")\
                                               .withColumnRenamed("Supervisor District", "SupervisorDistrict")

display(cleaned_sf_fire_calls_df)
```

```python
cleaned_sf_fire_calls_df.printSchema()
```

#### Q1. How many distinct typs of calls were made to the fire dept ?

```python
# By using countDistinct method 
q1_df = cleaned_sf_fire_calls_df.where("CallType is not null")\
                                .select(countDistinct("CallType"))

display(q1_df)

# Other method 
q1_df1 = cleaned_sf_fire_calls_df.where("CallType is not null")\
                                .select("CallType")\
                                .distinct()
print(q1_df1.count())
```

#### Q2. What are the different distinct types of calls made to the fire dipt ?

```python
q2_df = cleaned_sf_fire_calls_df.where("CallType is not null")\
                                .select("CallType")\
                                .distinct()

display(q2_df)
```

#### Q3. Find out all the responses or delayed times greater than 5 mins

```python
q3_df = cleaned_sf_fire_calls_df.select("CallNumber", "delay")\
                                .where("delay > 5")

display(q3_df)
```

#### Q4. What were the most common call types ?

```python
q4_df = cleaned_sf_fire_calls_df.where("CallType is not null")\
                                .groupBy("CallType")\
                                .count()\
                                .orderBy("count", ascending=False)
                                
display(q4_df)
```

#### Q5. What ZIP codes accounted for the most commom calls ?

```python
q5_df = cleaned_sf_fire_calls_df.where("CallType is not null")\
                                .groupBy("CallType", "Zipcode")\
                                .count()\
                                .orderBy("count", ascending=False)

display(q5_df)
```

#### Q6. What San Francisco neighborhoods are in the ZIP codes 94102 and 94103 ?

```python
q6_df = cleaned_sf_fire_calls_df.where("Zipcode == 94102 or Zipcode == 94102")\
                                .select("Neighborhood", "Zipcode")

display(q6_df)
```

#### Q7. What was the sum of all call alarms, average, min and max of the call response time ?

```python
q7_df = cleaned_sf_fire_calls_df.selectExpr("sum(NumAlarms) as sum_alrams", "avg(delay) as avg_delay", "min(delay) as min_delay", "max(delay) as max_delay")
                                
display(q7_df)                              
```

#### Q8. How many distinct years of data in CSV files ?

```python
q8_df = cleaned_sf_fire_calls_df.withColumn("years", year("CallDate"))\
                                .select("years").distinct()

display(q8_df)
```

#### Q9. What week of the year in 2018 had the most fire calls ?

```python
q9_df = cleaned_sf_fire_calls_df.withColumn("year", year("CallDate"))\
                                .where("year== 2018")\
                                .withColumn("week_of_year", weekofyear("CallDate"))\
                                .groupBy("week_of_year")\
                                .count()\
                                .select("week_of_year", "count")\
                                .orderBy("count", ascending=False)

display(q9_df)
```

#### Q10. What neighborhoods in San Francisco had the worst respinse time in 2018 ?

```python
q10_df = cleaned_sf_fire_calls_df.withColumn("year", year("CallDate"))\
                                 .where("year==2018")\
                                 .select("Neighborhood", "delay")\
                                 .orderBy("delay", ascending=False)

display(q10_df)
```
